Напишите программу, которая запрашивает у пользователя количество строк для вывода. В каждой строке должна быть последовательность чисел от 1 до текущего номера строки, где:
Четные числа должны быть умножены на 2.
Нечетные числа должны быть возведены в квадрат.
После этого программа должна вывести результат в консоль с помощью функции print(). Для вычисления результата необходимо использовать именованные функции и вложенные циклы.

Формат ввода

7

Формат вывода

1

1 4

1 4 9

1 4 9 8

1 4 9 8 25

1 4 9 8 25 12

1 4 9 8 25 12 49

Примечания
Обязательно наличие в решении хотя бы одной именованной функции, объявленной через def и хотя бы одного вложенного цикла. Вывод результата сделать с помощью функции print().

```python
def calculate_number(n):
    """Вычисляет значение числа в соответствии с правилами"""
    if n % 2 == 0:  # Четное число
        return n * 2
    else:  # Нечетное число
        return n * n  # n в квадрате

def print_sequence(rows):
    """Выводит последовательности чисел для заданного количества строк"""
    for i in range(1, rows + 1):
        for j in range(1, i + 1):
            result = calculate_number(j)
            print(result, end=' ')
        print()  # Переход на новую строку

def main(rows=None):
    """Основная функция программы

    Args:
        rows: количество строк (если None, запрашивается у пользователя)
    """
    try:
        if rows is None:
            rows = int(input("Введите количество строк для вывода: "))

        if rows <= 0:
            print("Количество строк должно быть положительным числом.")
            return

        print_sequence(rows)

    except ValueError:
        print("Ошибка: введите целое число.")

if __name__ == "__main__":
    main()
```

Напишите функцию, которая принимает два числа и возвращает большее из них. Используйте именованную функцию для сравнения чисел и именованную функцию высшего порядка для её применения.
Формат ввода
7
3
.Формат вывода
7
Примечания
Обязательно наличие в решении хотя бы двух именованных функций, объявленных через def, при этом хотя бы одна из них должна принимать другую функцию в качестве аргумента.

```python
def compare_numbers(a, b):
    """Сравнивает два числа и возвращает большее"""
    if a > b:
        return a
    else:
        return b

def apply_comparison(func, x, y):
    """Принимает функцию сравнения и два числа, применяет её"""
    return func(x, y)

def main():
    """Основная функция для получения ввода и вывода результата"""
    try:
        # Чтение двух чисел
        num1 = int(input().strip())
        num2 = int(input().strip())

        # Применение функции сравнения через функцию высшего порядка
        result = apply_comparison(compare_numbers, num1, num2)

        # Вывод результата
        print(result)

    except ValueError:
        print("Ошибка: введите целые числа.")

if __name__ == "__main__":
    main()
```

Напишите функцию, которая принимает число N и возвращает первое простое число больше N. Используйте анонимную функцию для проверки, является ли число простым, и функцию высшего порядка для его нахождения.

Формат ввода
5

Формат вывода
7

Примечания
Обязательно наличие в решении хотя бы одной именованной функции, объявленной через def и хотя бы одной анонимной функции, объявленной через lambda.

```python
def next_prime(N):
    # Анонимная функция для проверки, является ли число простым
    is_prime = lambda n: n > 1 and all(n % i != 0 for i in range(2, int(n**0.5) + 1))

    # Функция высшего порядка для нахождения первого простого числа больше N
    def find_next_prime(start, condition):
        current = start + 1
        while True:
            if condition(current):
                return current
            current += 1

    return find_next_prime(N, is_prime)
```

Напишите функцию, которая принимает длины двух катетов прямоугольного треугольника и возвращает длину гипотенузы. Используйте вложенную анонимную функцию для вычисления.

Формат ввода
3

4

Формат вывода
5

Примечания
Обязательно наличие в решении хотя бы одной именованной функции, объявленной через def и хотя бы одной анонимной функции, объявленной через lambda.

```python
def calculate_hypotenuse(a, b):
    """Принимает длины двух катетов и возвращает длину гипотенузы"""
    # Непосредственно возвращаем результат вызова анонимной функции
    return (lambda x, y: (x**2 + y**2)**0.5)(a, b)
```
